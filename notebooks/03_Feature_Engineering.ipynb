**Load dataset**

**Drop Irrelevant or High-Null Columns**

```python
# List of columns to drop
drop_cols = [
    # IDs, URLs
    'id', 'member_id', 'url', 'desc', 'title', 'zip_code',

    # Dates and post-loan info
    'issue_d', 'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d',
    'earliest_cr_line', 'sec_app_earliest_cr_line',

    # Hardship and settlement
    'hardship_flag', 'hardship_type', 'hardship_reason', 'hardship_status',
    'deferral_term', 'hardship_amount', 'hardship_start_date', 'hardship_end_date',
    'payment_plan_start_date', 'hardship_length', 'hardship_dpd',
    'hardship_loan_status', 'orig_projected_additional_accrued_interest',
    'hardship_payoff_balance_amount', 'hardship_last_payment_amount',
    'settlement_status', 'settlement_date', 'settlement_amount',
    'settlement_percentage', 'settlement_term', 'debt_settlement_flag',
    'debt_settlement_flag_date',

    # Co-applicant info
    'revol_bal_joint', 'annual_inc_joint', 'dti_joint', 'verification_status_joint',
    'sec_app_fico_range_low', 'sec_app_fico_range_high', 'sec_app_inq_last_6mths',
    'sec_app_mort_acc', 'sec_app_open_acc', 'sec_app_revol_util',
    'sec_app_open_act_il', 'sec_app_num_rev_accts',
    'sec_app_chargeoff_within_12_mths', 'sec_app_collections_12_mths_ex_med',
    'sec_app_mths_since_last_major_derog'
]

# Only drop columns that exist in the current DataFrame
drop_cols = [col for col in drop_cols if col in df.columns]
df.drop(columns=drop_cols, inplace=True)
print("Shape after dropping:", df.shape)
```

```python
# Show columns still having > 90% missing values
high_nulls = df.isnull().mean() * 100
high_nulls[high_nulls > 90].sort_values(ascending=False)
```

```python
df.info()
df.head()
```

```python
# Missing %
missing_percent = df.isnull().mean() * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)
missing_percent
```

**Impute Numerical Features**

```python
# Fill numeric columns with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
```

**Impute Categorical Features**

```python
# Fill categorical columns with mode
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
```

```python
# Check if anything remains null
print("Remaining nulls:", df.isnull().sum().sum())
```

**Encode Categorical Variables**

```python
# List of Categorical Columns
df.select_dtypes(include='object').nunique().sort_values()
```

```python
# Encode Using One-Hot
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
# Preview Encoded Data
print("Encoded shape:", df_encoded.shape)
df_encoded.head()
```

```python
df_encoded.dtypes.value_counts()
```

```python
# Convert all bool columns to int (0/1)
df_encoded = df_encoded.astype({col: 'int' for col in df_encoded.select_dtypes('bool').columns})
```
