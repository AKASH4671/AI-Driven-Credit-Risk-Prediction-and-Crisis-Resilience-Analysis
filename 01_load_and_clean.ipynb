***Clean and Encode***

**All Unique Status Values**

```python
df['loan_status'].value_counts()
```

**Keep Only Completed Loans**

```python
# Keep only rows where loan status is 'Fully Paid', 'Charged Off', or 'Default'
df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off', 'Default'])]
```

**Create Binary Target Column: "loan_default"**

```python
# Encode target variable: 0 = Fully Paid, 1 = Default/Charged Off
df['loan_default'] = df['loan_status'].apply(lambda x: 1 if x in ['Charged Off', 'Default'] else 0)

# Drop original column
df.drop(columns=['loan_status'], inplace=True)

# Check new class balance
df['loan_default'].value_counts(normalize=True)
```

**Print Summary**

```python
print("Shape after filtering:", df.shape)
print("Default rate in dataset: {:.2f}%".format(df['loan_default'].mean() * 100))
```
